```python
from random import randint

# De la tabla de instancias M, escogemos los siguientes valores, respectivamente.
lista_meses = [7,15,25,35,45,115,145,175,185,195,500,650,700,800,900]
#inputs del costo de produccion en horario normal
c_inf = 5000
c_sup = 10000
#inputs del costo de produccion en horario extra
e_inf = 5000
e_sup = 10000
#inputs del costo de almacenamiento por unidad en el mes (valor fijo).
a_inf = 10
a_sup = 1000
#inputs capacidad de maxima de almacenamiento en el mes (valor fijo).
al_inf = 1
al_sup = 5000
#inputs de produccion normal de unidades en el mes
p_inf = 10000
p_sup = 20000

#Listas con las demandas de prueba
d=[]
#Listas con los costos de produccion en horario normal en el mes
c=[]
#Listas con los costo de produccion en horario extra en el mes
e=[]
#Lista costo de almacenamiento por unidad en el mes (Valor fijo)
a=[]
#Lista de capacidad de almacenamiento de unidades en el mes (Valor fijo)
al=[]
#Lista de maximos de produccion normal de unidades en el mes (Valor fijo)
p=[]
#Lista de maximos ded produccion extra de unidades en el mes (Valor fijo)
pe=[]

i = 0
while (i<15): # Repetir para las 15 instancias
  meses = lista_meses[i] # Cantidad de meses de la instancia
  
  CNmax = randint(p_inf,p_sup) # Parametro de produccion maxima en horario normal
  p.append([CNmax])

  CEmax = randint(int(CNmax/2),20000) # Parametro de produccion maxima en horario extra con restriccion de al menos la mitad de produccion en horario normal
  pe.append([CEmax])

  CAmax = randint(al_inf,al_sup) # Parametro de almacenamiento maximo
  al.append([CAmax])

  almacenamiento = randint(a_inf,a_sup)
  a.append([almacenamiento])
  print(i+1," CNmax : ",CNmax)
  print(i+1," CEmax : ",CEmax)
  print(i+1," CAmax : ",CAmax)

  i_2 = 0
  aux = 0
  lista_aux=[]
  while(i_2<meses): # Parametros costo de produccion en horario normal en el mes.
    if(i_2 == 0):
      superior = 1 - (meses/c_sup)
      valor = randint(c_inf,int(c_sup*superior))
    if(i_2 >= 0):
      aux_2 = c_sup-valor
      superior = 1 - ((meses+aux)/aux_2)
      valor = randint(valor+1,int((aux_2*superior)+valor))
    lista_aux.append(valor)
    aux-=1
    i_2+=1
  c.append(lista_aux)

  i_2 = 0
  lista_aux = []
  while (i_2<meses): # Parametros costo de produccion en horario extra en el mes.
    flag = True
    while flag:
      valor = randint(e_inf,e_sup)
      if(i_2 == 0):
        lista_aux.append(valor)
        flag = False
      if(i_2 >= 1):
        flag = valor in lista_aux
        if(not(flag)):
          lista_aux.append(valor)
    i_2+=1
  e.append(lista_aux)

  lista_aux = []
  i_2 = 0
  aux = 0
  almacenado = 0
  dif= 0
  valor=0
  if CAmax > CNmax + CEmax:
    CAmax = CNmax + CEmax - 1
  aux_meses = meses + 5000
  while (i_2<meses): # Parametros de demanda, cumpliendo el requisito de cada demanda mayor al paso del mes
    if(i_2 == 0):
      d_max = CNmax + CEmax + CAmax
      if (aux_meses>CAmax):
        valor = randint(aux_meses,d_max)
      else:
        valor = randint(CAmax,d_max)
      valor = randint(d_max - CAmax,d_max)
      if valor > CNmax + CEmax:
        almacenado =  valor - CNmax - CEmax
        dif = CAmax - almacenado
    else:
      if dif > 0 and aux_meses>1:
        if dif > almacenado:
          d_max = CNmax + CEmax + almacenado - 1
          if d_max > valor:
            d_max = valor
          d_inf = d_max - (2*almacenado)
        else:
          d_max = CNmax + CEmax + dif - 1
          if d_max > valor:
            d_max = valor
          d_inf = d_max - (2*dif)
        if (aux_meses > d_inf):
          d_inf = aux_meses
        valor = randint(d_inf,d_max)
        if valor > CNmax + CEmax:
          almacenado +=  valor - CNmax - CEmax
          dif = CAmax - almacenado
        else:
          d_max-=valor
          almacenado -=  d_max
          dif = 0
      elif almacenado >0 and aux_meses>1:

        d_max = valor-1
        valor = randint(aux_meses,d_max)
        d_max -= valor
        almacenado -=  d_max
      elif almacenado >0 and aux_meses==1:
        d_max = valor - almacenado
        valor = randint(aux_meses,d_max)
        almacenado = 0
      else:
        d_max = valor-1
        valor = randint(aux_meses,d_max)
    lista_aux.append(valor)
    aux_meses-=1
    i_2+=1

  lista_aux.reverse()
  d.append(lista_aux)

  print("")
  print("Valores de los parametros para una instancia de ",lista_meses[i]," meses")
  print("Para la demanda es: ",d)
  print("Para el costo de produccion horario normal con un rango de [",c_inf,",",c_sup,"] es: ",c)
  print("Para el costo de produccion horario extra con un rango de [",e_inf,",",e_sup,"] es: ",e)
  print("Para el costo de almacenamiento con un rango de [",a_inf,",",a_sup,"] es: ",a)
  print("Para la capacidad de almacenamiento con un rango de [",al_inf,",",al_sup,"] es: ",al)
  print("Para el maximo de produccion normal con un rango de [",p_inf,",",p_sup,"] es: ",p)
  print("Para el maximo de produccion extra con un rango de [",CNmax/2,",",20000,"] es: ",pe)
  print("")


  txt = 'int: meses = '+ str(lista_meses[i]) +';\n'
  txt += 'int: CNmax = '+ str(CNmax) +';\n'
  txt += 'int: CEmax = '+ str(CEmax) +';\n'
  txt += 'int: CAmax = '+ str(CAmax) +';\n\n'
  txt += 'array[1..meses] of int: d;\n'
  txt += 'd=['
  for i_3 in range(meses):
    if i_3 < meses-1: 
      txt+= str(d[0][i_3])+', '
    else:
      txt+= str(d[0][i_3])+'];\n\n'

  txt+= 'int: M = sum(i in 1..meses)(d[i]);\n\n'

  txt+='array[1..meses] of var 0..1: x;\n'
  txt+='array[1..meses] of var int: y;\n'
  txt+='array[1..meses] of var int: h;\n'
  txt+='array[1..meses] of var int: p;\n'
  txt+='var int: ct;\n'

  #Funcion objetivo
  fo = 'constraint ct = '
  for t in range(meses):
    fo += str(c[0][t]) + '*p[' + str(t+1)+ '] + '
    fo += str(e[0][t]) + '*h[' + str(t+1)+ '] + '
    fo += str(a[0][0]) + '*y[' + str(t+1)
    if t < meses-1:
      fo+= '] + '
    else:
      fo+= '];\n\n'

  txt+=fo
  
  txt+='constraint p[1]+ h[1] <= 0.2*sum(i in 1..meses)(p[i]+h[i]);\n'
  txt+='constraint forall(i in 1..meses)(h[i] >= 0.5*p[i] - M*(1-x[i]));\n'
  txt+='constraint forall(i in 1..meses)(h[i] <= M*x[i]);\n'
  txt+='constraint forall(i in 1..meses)(y[i] <= CAmax);\n'
  txt+='constraint y[1] = p[1] + h[1] - d[1];\n'
  txt+='constraint forall(i in 2..meses)(y[i] = y[i-1] + p[i] + h[i] - d[i]);\n'
  txt+='constraint y[meses] = 0;\n'
  txt+='constraint p[1] + h[1] >= d[1];\n'
  txt+='constraint forall(i in 2..meses)(p[i] + h[i] + y[i-1] >= d[i]);\n'
  txt+='constraint forall(i in 1..meses)(p[i] >= 0);\n'
  txt+='constraint forall(i in 1..meses)(h[i] >= 0);\n'
  txt+='constraint forall(i in 1..meses)(y[i] >= 0);\n'
  txt+='constraint forall(i in 1..meses)(CNmax >= p[i]);\n'
  txt+='constraint forall(i in 1..meses)(CEmax >= h[i]);\n\n'

  txt+='solve minimize ct;'

  mzn = open('modelo_minizinc '+str(i+1)+'.mzn','w')
  mzn.write(txt)
  mzn.close()


  d=[]
  c=[]
  e=[]
  a=[]
  al=[]
  p=[]
  pe=[]
  i+=1
```
