## Enunciado

Los sistemas de bicicletas compartidas son una forma de alquilar bicicletas en las que el proceso de obtención de membresía, alquiler y devolución de bicicletas se automatiza a través de una red de quioscos en toda la ciudad. Con estos sistemas, las personas pueden alquilar una bicicleta en un lugar y devolverla en un lugar diferente según sea necesario. Actualmente, hay más de 500 programas de bicicletas compartidas en todo el mundo. Los datos generados por estos sistemas los hacen atractivos para los investigadores porque la duración del viaje, el lugar de salida, el lugar de llegada y el tiempo transcurrido se registran explícitamente. Por lo tanto, los sistemas de bicicletas compartidas funcionan como una red de sensores que se puede utilizar para estudiar la movilidad en una ciudad. En esta taller, se les muestra a kis participantes el alquiler de bicicletas en el programa Capital Bikeshare en Washington, D.C.

## Pregunta 1

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#Dataframes
datos = pd.read_csv("bike.csv", header = 0)

#Para el Gráfico 1 
etiqueta1 = ["0", "1"]

#Workingday 0/1
datosworkingday0 = datos[datos["workingday"] == 0]
datosworkingday1 = datos[datos["workingday"] == 1]

#Cantidad de alquileres por 0/1
c0 = datosworkingday0["count"].sum()
c1 = datosworkingday1["count"].sum()

#Gráfico 1
fig, ax = plt.subplots()
co = np.arange(2)
an = 0.30
ax.bar([0,1], [c0,c1] , an, ec = "black")
ax.set_title("Gráfico 1")
ax.set_xticks(co)
ax.set_xticklabels(etiqueta1)
ax.set_xlabel("Alquileres", fontdict = {'fontsize':10})
ax.set_ylabel("Workingday", fontdict = {'fontsize':10})

plt.show()
```

Dada la magnitud de las barras se evidencia claramente que la opción 1 es la más demandada, lo que significa que la gente prefiere alquilar bicicletas para movilizarse a sus trabajos y hogares. De esto podríamos concluir que la gente desea ahorrar combustible que es un gasto notorio en el bolsillo de las personas y también se optimiza su tiempo ejercitando antes de ir a su destino  y los días que no trabajan prefieren estar en la comodidad de su hogar.


## Pregunta 2

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#Dataframes
datos = pd.read_csv("bike.csv", header = 0)

#Para el Gráfico 2 
etiqueta2 = ["Invierno", "Primavera", "Verano", "Otoño"]

#Años 2011/2012
datos2011 = datos[datos["datetime"] < "2012-01-01 00:00:00"]
datos2012 = datos[datos["datetime"] >= "2012-01-01 00:00:00"]

#Seasons Invierno/Primavera/Verano/Otoño | 2011/2012         
datos20111 = datos2011[datos2011["season"] == 1] 
datos20112 = datos2011[datos2011["season"] == 2] 
datos20113 = datos2011[datos2011["season"] == 3] 
datos20114 = datos2011[datos2011["season"] == 4] 

datos20121 = datos2012[datos2012["season"] == 1]
datos20122 = datos2012[datos2012["season"] == 2]
datos20123 = datos2012[datos2012["season"] == 3]
datos20124 = datos2012[datos2012["season"] == 4]

#Cantidad de alquileres por seasons Primavera/Verano/Otoño/Invierno
c20111 = datos20111["count"].sum()
c20112 = datos20112["count"].sum()
c20113 = datos20113["count"].sum()
c20114 = datos20114["count"].sum()
valores2011 = [ c20111, c20112, c20113,c20114]

c20121 = datos20121["count"].sum()
c20122 = datos20122["count"].sum()
c20123 = datos20123["count"].sum()
c20124 = datos20124["count"].sum()
valores2012 = [ c20121, c20122, c20123,c20124]

#Grafico 2
fig, ax1 = plt.subplots()
co = np.arange(4)
an = 0.30

ax1.bar(co - an/2, valores2011, an, ec = "black", label = "2011")
ax1.bar(co + an/2, valores2012, an, ec = "black", label = "2012")
ax1.set_title("Gráfico 2")
ax1.set_xticks(co)
ax1.set_xticklabels(etiqueta2)
ax1.set_xlabel("Seasons", fontdict = {'fontsize':10})
ax1.set_ylabel("Cant de Alquileres", fontdict = {'fontsize':10})
plt.legend()
plt.show()
```

En el gráfico se pueden apreciar las barras azules y naranjas, que corresponden a los años 2011 y 2012 respectivamente. Notoriamente se puede determinar que hubo un aumento en el alquiler de bicicletas en el año 2012, siendo la temporada de invierno la que posee un aumento más considerable entre las 4. Cabe señalar que el ciclo en cada año fue sostenido. De esto podemos concluir que el negocio de alquileres de bicicletas tuvo un éxito luego del primer año.

## Pregunta 3

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas import plotting

%matplotlib inline
#leer el archivo csv
data = pd.read_csv("bike.csv")

#creacion del boxplot a partir del archivo csv, en la funcion se determina las variables de los ejes, el tamaño del grafico, y los colores de sus propiedad3es
ax = data.plot.box(column=['temp'], by=['season'], figsize=(8, 6), patch_artist = True, 
                   boxprops=dict(facecolor='lightcyan', color='lightslategray'),
                   capprops=dict(color='black'),
                   whiskerprops=dict(color='black'),
                   flierprops=dict(color='black', markeredgecolor='black'),
                   medianprops=dict(color='black'))

#titulo
plt.title('Boxplot Temperatura registrada por cada Season')

#nombres de los ejes
plt.xlabel("Season")
plt.ylabel("Temperatura")

#lineas de referencia
plt.grid(True)

plt.show()
```

Claramente se observa que las mayores temperaturas se alcanzaron en la estación de verano con una poca dispersión de sus datos, en otras palabras, la temperatura de los días que se alquilaron bicicletas no cambio tanto durante esta estación del año, en relación a las otras estaciones la dispersión de sus datos es mayor y en invierno y otoño las temperaturas entre 50% y el 75% de todas las temperaturas registradas los días que se realizaron los alquileres son más dispersas que las que están entre el 25% y el 50%, es decir las temperaturas varían más a medida que subían en estas estaciones.

## Pregunta 4

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas import plotting

%matplotlib inline
data = pd.read_csv("bike.csv")

#agrupacion de la cant total de alquileres de cada temporada
total_count = data.groupby('season')['count'].sum()

#lista de cada temporda [1,2,3,4]
s = list(data['season'].value_counts().index)

#figura donde va el grafico
fig, ax1 = plt.subplots(figsize = (6,4))

#grafico de barra, eje x, eje y , colores de las barras
graf = ax1.bar(sorted(s), total_count,color=['lightblue','lightpink','lightgreen','gold'])

#titulo
plt.title("Count by Season", fontdict ={'fontsize':20})

#nombre de los ejes
plt.ylabel("Counts", fontdict ={'fontsize':15})
plt.xlabel("Season", fontdict ={'fontsize':15})
```

Se observa que la estación del año en que más personas alquilan bicicletas es verano seguida de primavera y que a partir de la comparación de temperaturas del grafico anterior mientras más frio menos personas alquilan bicicletas y de la misma manera mientras más alta la temperatura más personas las alquilan.

## Pregunta 5

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings

data_set = pd.read_csv('bike.csv')

data = data_set[data_set['season'] == 4]
data = data[['count','weather']]    #se filtra el dataframe original para obtener uno que tenga solo la informacion necesaria aqui

prom_w1 = data[data['weather'] == 1]
prom_w1 = prom_w1.mean()
prom_w1 = prom_w1['count'] #toma el valor del promedio de rentas cuando el clima es de tipo 1, se hace lo mismo para cada clima

prom_w2 = data[data['weather'] == 2]
prom_w2 = prom_w2.mean()
prom_w2 = prom_w2['count']

prom_w3 = data[data['weather'] == 3]
prom_w3 = prom_w3.mean()
prom_w3 = prom_w3['count']

prom_w4 = data[data['weather'] == 4]
prom_w4 = prom_w4.mean()
prom_w4 = prom_w4['count']


valores_x = [1,2,3,4]
valores_y = [prom_w1, prom_w2, prom_w3, prom_w4]

co = np.arange(len(valores_x))
an = 0.3

fig, ax = plt.subplots()
ax.bar(co,valores_y,an)
ax.set_xticks(co)
ax.set_xticklabels(valores_x)
ax.set_xlabel("weather type")
ax.set_ylabel("mean of users")
```

Observando el grafico es clara la baja o nula cantidad de alquileres cuando se experimenta un clima del tipo 3 o 4, que corresponde a días donde hay lluvia y o nieve, cosa que es de esperarse dadas las dificultades que implica ir en bicicleta en dichas condiciones.

Por otro lado vemos una alza cuando el clima es del tipo 1 o 2, que corresponde a dias donde las nubes son escasas y está mayormente soleado, lo que sugiere que el uso de las bicicletas tiene un objetivo más recreacional que de utilidad (como para movilizarse al trabajo por ejemplo), cosa que contrasta con lo visto en la pregunta 1 y 7. Una posible respuesta a esta inconcordancia puede estar relacionada a que este grafico es especifico para la estacion de otoño, donde el calor deja de ser extremo a diferencia del verano, lo que podria causar un aumento en el aquiler recreacional de bicicletas.

## Pregunta 6

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#Dataframes
datos = pd.read_csv("bike.csv", header = 0)

#Grafico 6 

#Invierno
invierno = datos[datos["season"] == 4] 
min1 = invierno["atemp"].min()
max1 = invierno["atemp"].max()
k = 14 #calcular k con teorema de sturges
r = max1 - min1
a = r/k #amplitud
co61 = np.arange(0,35,round(a))

#Verano
verano = datos[datos["season"] == 2]
min2 = invierno["atemp"].min()
max2 = invierno["atemp"].max()
k = 13 #calcular k con teorema de sturges
r = max2 - min2
a = r/k #amplitud
co62 = np.arange(0,45,round(a))

#Grafico 6 invierno
fig, ax61 = plt.subplots()
ax61.hist(invierno["atemp"], 13, ec = "black")
ax61.set_xticks(co61)

ax61.set_title('Invierno', fontdict = {'fontsize':14})
ax61.set_xlabel("Temperatura aparente en Invierno", fontdict = {'fontsize':10})
ax61.set_ylabel("Cant de Alquileres", fontdict = {'fontsize':10})

#Grafico 6 verano
fig, ax62 = plt.subplots()
ax62.hist(verano["atemp"], 16, ec = "black")
ax62.set_xticks(co62)

ax62.set_title('Verano', fontdict = {'fontsize':14})
ax62.set_xlabel("Temperatura aparente en Verano", fontdict = {'fontsize':10})
ax62.set_ylabel("Cant de Alquileres", fontdict = {'fontsize':10})

plt.show()
```

Se puede observar que en invierno la gente tiende a alquilar bicicletas a una temperatura aparente de 16 a 18 grados, por otro lado en verano la gente prefiere alquilar a una temperatura aparente de 26 a 32 grados, siendo éstos la moda local de cada histograma. 
Podemos concluir que en invierno la gente prefiere alquilar bicicletas a una temperatura más elevada debido al frío de invierno, en el caso del histograma de verano visualmente se puede concluir que la moda local está lejos de la máxima temperatura señalada, por lo que la gente prefiere alquilar bicicletas a una temperatura menor.

## Pregunta 8

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings

data_set = pd.read_csv('bike.csv')

casuals = data_set['casual']
casuals = casuals.sum() #es la cantidad total de usuarios no registrados

registered = data_set['registered']
registered = registered.sum()   #es la cantidad total de usuarios registrados

valores_y = [registered, casuals]
valores_x = ["registered","casuals"]

co = np.arange(len(valores_x))
an = 0.3

fig, ax = plt.subplots()
ax.bar(co,valores_y,an)
ax.set_xticks(co)
ax.set_xticklabels(valores_x)
ax.set_xlabel("Total de alquileres")
ax.set_ylabel("Tipo de usuario")
```

Este último gráfico se confeccionó con intencion de reforzar las conclusiones planteadas para la pregunta 1 y 7, donde se sugiere que el uso de las bicicletas está más relacionado con ser un medio de movilización (ya sea para el trabajo o salidas personales) que para recreación.
